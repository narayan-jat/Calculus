This is the program to plot graph for analytical and numerical solution for double differential equation. Differential equation is:

**d^2x/dy^2 = Ï‰^2x**

```python
# This cell imports necessary modules required for plottings.
import matplotlib.pyplot as px
import math as m
```

```python
# This cell solves double differential equation by analytical approach.
a_x_list = []
t_list = []
def analytical(time, omega):
    '''
    Returns the ploting for analytical solution of Double differential equation.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.

    parameter omega: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global a_x_list
    global t_list
    a_x_list = []
    time = int(1 // time)
    t_list = [t / time for t in range(1, 5 * time + 1)]  # time list created with considering given time difference.
    for t in t_list:
        x_t = ((omega + 1) * m.exp(omega * t) + (omega - 1) * m.exp(-omega * t)) / 2 * omega
        a_x_list.append(x_t)  # Crating list for analytical solution.
    px.plot(t_list, a_x_list, label = 'Analytical solution')
    px.legend()
```

```python
# This cell creates list of time values ans list of values received after applying Euler method.
num_x_list_euler = []
def numericalEuler(time, omega):
    '''
    Returns the ploting for numerical solution of Double differential equation by Euler method.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.

    parameter omega: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global num_x_list_euler
    delta_t = time
    time = int(1 // time)
    num_x_list_euler = []  # already added value at x(0).
    x_t_at_0 = 1  # Value of x at t = 0.
    v_t = 1  # Value of v at t = 0.
    for i in range(5 * time):
        x_t = v_t * delta_t + x_t_at_0      # calculating next values by using previous calculated one.
        v_t = (omega ** 2) * (x_t_at_0 * delta_t) + v_t
        num_x_list_euler.append(x_t)
        x_t_at_0 = x_t   # changing older value with newer one.
    px.plot(t_list, num_x_list_euler, label = 'Numerical by Euler Method')
    px.legend()
```

```python
# This cell creates list of time values ans list of values received after applying Mid point method.
num_x_list_mid = []
# num_t_list = []
def numericalmid(time, omega):
    '''
    Returns the ploting for numerical solution of Double differential equation by Mid Point Method method.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.

    parameter omega: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global num_x_list_mid
    delta_t = time
    time = int(1 // time)
    num_x_list_mid = []  # already added value at x(0).
    x_t_at_0 = 1
    v_t_at_0 = 1
    initial_t = 0   # Initial value of t
    for i in range(5 * time):
        x_t = x_t_at_0 +  delta_t * (x_t_at_0 + 0.5 * delta_t * v_t_at_0)   # creating middle value of x
        initial_t += delta_t
        v_t = v_t_at_0 + delta_t * (v_t_at_0 + 0.5 * delta_t * (omega ** 2 * x_t_at_0))         # finding values for v
        # Reassigning new calculated values.
        x_t_at_0 = x_t
        v_t_at_0 = v_t
        num_x_list_mid.append(x_t)
    px.plot(t_list, num_x_list_mid, label = 'Numerical by Mid Point Method')
    px.legend()
```

```python
# Graph for error function for Mid Point Method.
def errorMid():
    errorList = []   # List accumulator to contain values for plotting error graph.
    for i in range(len(a_x_list)):
        errorExp = abs((a_x_list[i] - num_x_list_mid[i]) / a_x_list[i])  # values for error expression.
        errorList.append(errorExp)
    px.plot(t_list, errorList, label = 'Mid Point method Error')
    px.legend()
```

```python
# Graph error function for Euler Method.
def errorEuler():
    errorList = []   # List accumulator to contain values for plotting error graph.
    for i in range(len(a_x_list)):
        errorExp = abs((a_x_list[i] - num_x_list_euler[i]) / a_x_list[i])  # values for error expression.
        errorList.append(errorExp)
    px.plot(t_list, errorList, label = 'Euler Error')
    px.legend()
```
