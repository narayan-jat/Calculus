This is the program to solve ODE using **Mid Point Method**. There are plotting of analytical solution, numerical solution(By mid point method) and error function.
ODE is:
> **dx/dt = x**

```python
import matplotlib.pyplot as px
import plotly.graph_objects as go
import math as m
```

```python
# time list and value list for e^t(Analytical method)
a_x_list = []
a_t_list = []
def analytical(time):
    '''
    Returns the ploting for analytical solution of O.D.E.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global a_x_list
    time = int(1 // time)
    a_t_list = [t / time for t in range(5 * time + 1)]  # time list created with considering given time difference.
    a_x_list = [eval('m.exp(t)') for t in a_t_list]  # Evaluating using exp function.
    px.plot(a_t_list, a_x_list, label = 'Analytical')
    px.legend()
```

```python
# time list ans value list for descrete value  By Mid Point Method.
d_x_list = []
d_t_list = []
def numerical(time):
    '''
    Returns the ploting for numerical solution of O.D.E.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global d_x_list
    global d_t_list
    delta_t = time
    time = int(1 // time)
    # creating time list with given time ranges.
    d_t_list = [t / time for t in range(5 * time + 1)]
    d_x_list = [1]  # already added value at x(0).
    x_t = 1
    initial_t = 0
    for i in range(5 * time):
        t_mid = initial_t + 0.5 * delta_t  # creating middle value of t
        x_mid = x_t + 0.5 * delta_t * (x_t)   # creating middle value of x
        initial_t += delta_t
        x_t += delta_t * x_mid    # finally calculating x_n+1.
        d_x_list.append(x_t)
    px.plot(d_t_list, d_x_list, label = 'Numerical')
    px.legend()
```

```python
# Graph for error function.
def error():
    errorList = []   # List accumulator to contain values for plotting error graph.
    for i in range(len(a_x_list)):
        errorExp = abs((a_x_list[i] - d_x_list[i]) / a_x_list[i])  # values for error expression.
        errorList.append(errorExp)
    px.plot(d_t_list, errorList, label = 'Error')
    px.legend()
```

```python
# Now calling functions to plot graph with lowering delta_t value.
print('Graph for delta_t 0.1 is: ')
numerical(0.1)
analytical(0.1)
px.show()
print('Graph showing plotting for error occured:  ')
error()
px.show()
print('Graph for delta_t 0.01 is: ')
numerical(0.01)
analytical(0.01)
px.show()
print('Graph showing plotting for error occured:  ')
error()
px.show()
print('Graph for delta_t 0.001 is: ')
numerical(0.001)
analytical(0.001)
px.show()
print('Graph showing plotting for error occured:  ')
error()
px.show()
```
