This is the program to solve O.D.E using **Mid Point Method**. There are plotting of analytical solution, numerical solution(By mid point method) and error function.
ODE is:
> **dx/dt = x**

```python
# This cell imports necessary modules required for plottings.
import matplotlib.pyplot as px
import plotly.graph_objects as go
import math as m
```

```python
# This cell creates list of time and list of values received after applying Anaytical method.
a_x_list = []
def analytical(time):
    '''
    Returns the ploting for analytical solution of O.D.E.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global a_x_list
    time = int(1 // time)
    a_t_list = [t / time for t in range(5 * time + 1)]  # time list created with considering given time difference.
    a_x_list = [eval('m.exp(t)') for t in a_t_list]  # Evaluating using exp function.
    px.plot(a_t_list, a_x_list, label = 'Analytical')
    px.legend()
```

```python
# This cell creates list of time values ans list of values received after applying Mid point method.
num_x_list_mid = []
num_t_list = []
def numericalmid(time):
    '''
    Returns the ploting for numerical solution of O.D.E.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global num_x_list_mid
    global num_t_list
    delta_t = time
    time = int(1 // time)
    # creating time list with given time ranges.
    num_t_list = [t / time for t in range(5 * time + 1)]
    num_x_list_mid = [1]  # already added value at x(0).
    x_t = 1
    initial_t = 0   # Initial value of t
    for i in range(5 * time):
        t_mid = initial_t + 0.5 * delta_t  # creating middle value of t
        x_mid = x_t + 0.5 * delta_t * (x_t)   # creating middle value of x
        initial_t += delta_t
        x_t += delta_t * x_mid    # finally calculating x_n+1.
        num_x_list_mid.append(x_t)
    px.plot(num_t_list, num_x_list_mid, label = 'Numerical by Mid Point Method')
    px.legend()
```

```python
# Graph for error function for Mid Point Method.
def errorMid():
    errorList = []   # List accumulator to contain values for plotting error graph.
    for i in range(len(a_x_list)):
        errorExp = abs((a_x_list[i] - num_x_list_mid[i]) / a_x_list[i])  # values for error expression.
        errorList.append(errorExp)
    px.plot(num_t_list, errorList, label = 'Mid Point method Error')
    px.legend()
```

This is the program to plot graphs for analytical solution, numerical solution(by Euler method) and error function.
The ODE is:

> **dx/dt = x**



```python
# This cell creates list of time values ans list of values received after applying Euler method.
num_x_list_euler = []
def numericaleuler(time):
    '''
    Returns the ploting for numerical solution of O.D.E.

    parameter time: It is the time difference at which graph to be plotted.
    precondition: It must be numeric.
    '''
    global num_x_list_euler
    delta_t = time
    time = int(1 // time)
    num_x_list_euler = [1]  # already added value at x(0).
    x_t = 1   # Value of x at t = 0.
    for i in range(5 * time):
        x_tplus_1 = x_t * delta_t + x_t  # calculating next values by using previous calculated one.
        num_x_list_euler.append(x_tplus_1)
        x_t = x_tplus_1   # changing older value with newer one.
    px.plot(num_t_list, num_x_list_euler, label = 'Numerical by Euler Method')
    px.legend()
```

```python
# Graph error function for Euler Method.
def errorEuler():
    errorList = []   # List accumulator to contain values for plotting error graph.
    for i in range(len(a_x_list)):
        errorExp = abs((a_x_list[i] - num_x_list_euler[i]) / a_x_list[i])  # values for error expression.
        errorList.append(errorExp)
    px.plot(num_t_list, errorList, label = 'Euler Error')
    px.legend()
```

```python
# Now calling functions to plot graph with lowering delta_t value.
print('Graph for Mid Point Method delta_t 0.1 is: ')
numericalmid(0.1)
analytical(0.1)
px.show()
print('Graph for Euler Method delta_t 0.1 is: ')
numericaleuler(0.1)
analytical(0.1)
px.show()
print("Graph showing error relation between mid point method and euler method for given delta: ")
errorEuler()
errorMid()
px.show()
print('Graph for Mid Point Method delta_t 0.01 is: ')
numericalmid(0.01)
analytical(0.01)
px.show()
print('Graph for Euler Method delta_t 0.01 is: ')
numericaleuler(0.01)
analytical(0.01)
px.show()
print("Graph showing error relation between mid point method and euler method for given delta: ")
errorEuler()
errorMid()
px.show()
print('Graph for Mid Point Method delta_t 0.001 is: ')
numericalmid(0.001)
analytical(0.001)
px.show()
print('Graph for Euler Method delta_t 0.001 is: ')
numericaleuler(0.001)
analytical(0.001)
px.show()
print("Graph showing error relation between mid point method and euler method for given delta: ")
errorEuler()
errorMid()
px.show()
```
