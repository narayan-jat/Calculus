This module is dedicated to learn numerical method of integration when mathematical intergration is not possible. We will use numerical method of integration to calculate arc length of ellipse. Equation of ellipse is:
$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$$


```python
# Importing required module.
from scipy.integrate import quad
import math
import matplotlib.pyplot as py
from mpl_toolkits import mplot3d
```

```python
# difining ellipse equation.
def arcLength(t):
    '''
    '''
    return math.sqrt(a ** 2 * math.sin(t) ** 2 + b ** 2 * math.cos(t) ** 2)
```

```python
# Implementation of numerical integration.
aList = []
bList = []
intList = []
a = 2.0     # setting a = 2.0 intitially.
while a < 4:
    b = 1.0     # setting b = 1.0 intitially.
    while b < 2.0:
        intList.append(2 * quad(arcLength, 0, math.pi)[0])
        aList.append(a)
        bList.append(b)
        b += 0.1
    a += 0.1
```

```python
# Creating figure and axes
fig = py.figure()
ax = py.axes(projection='3d')

# Plot data
ax.plot3D(aList, bList, intList, 'red')

# Add labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot')

# Show plot
py.show()
```

# Shrinivas Ramanujan approximate formula for Arc length of Ellipse.
arc length of Ellipse:
$$π\left(3(a + b) - \sqrt{(3a + b)(3b + a)} \right ) \textit{where a and b are major and minor semi axis of ellipse respectively.}$$
This is an approximate expression which gives less accuracy. Following is the more accurate expression:
$$π\left((a + b)(1 + \frac{3h}{10 + \sqrt{4 - 3h}}) \right ) \textit{where a and b are major and minor semi axis of ellipse respectively and $ h = (\frac{a - b}{a + b})^{2}$}$$

```python
arcLength = [] 
for i in range(200):
  a = aList[i]
  b = bList[i]
  h = ((a-b)/(a+b))**2
  res = math.pi*(a+b)*(1+3*h/(10+(4-3*h)**0.5))
  arcLength.append(res)
```

```python
# Creating figure and axes
fig = py.figure()
ax = py.axes(projection='3d')

# Plot data
ax.plot3D(aList, bList, arcLength, 'violet')

# Add labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot')

# Show plot
py.show()
```
